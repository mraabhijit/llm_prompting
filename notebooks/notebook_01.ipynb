```python
from transformers import pipeline, AutoTokenizer
```

```python
import torch
```

```python
torch.manual_seed(0)
```

```python
model = 'tiiuae/falcon-7b-instruct'
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
pipe = pipeline(
    'text-generation',
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    device_map='auto',
)
```

```python
TEXT = """Directed by Jeff Wadlow and co-written with Greg Erb and Jason Oremland, ‘Imaginary’ has a premise filled with potential, but its execution leaves much to be desired. The narrative, involving a family grappling with change and sinister events tied to a stuffed bear named Chauncey, aims to delve into complex family tension and trauma themes. However, the film falters significantly in its development, relying on mainstream horror clichés and an underdeveloped plot that feels both convoluted and uninspired. The disjointed nature of the screenplay suggests there were too many cooks in the writing room, leaving us with a tedious narrative that strays far from imagination and becomes confusing. Using a teddy bear named Chauncey as the vessel for the unfolding horror is a concept that, while initially promising, quickly becomes another missed opportunity. The bear's connection to the family's past and its role in a dangerous scavenger hunt could have provided a unique avenue for exploring childhood fears and the power of imagination. Instead, the film prolongs the mystery to the point of frustration, culminating in a finale that feels both overextended and underwhelming. Moreover, the visual representation of Chauncey and other supernatural elements lacks originality and effectiveness. The reliance on practical effects and CGI does little to salvage the film's horror elements, resulting in a final product that is neither frightening nor engaging. The central characters—siblings Taylor (Taegen Burns) and Alice (Pyper Braun), stepmother Jessica (DeWanda Wise), and their absentee father Max—are caught in a whirlwind of poorly explained motivations and narrative shortcuts that do more to obscure than make their struggles relatable. Despite the commendable performances by the actors, ‘Imaginary’ fails to establish coherence and emotional resonance. The film's attempt to comment on the sacrificial nature of parenting and the horrors of mental illness gets lost in a messy script that veers unexpectedly into a dark fantasy realm, leaving audiences questioning the purpose and direction of the story. Despite the efforts of its cast, the film ultimately succumbs to a confusing plot, underdeveloped themes, and an over-reliance on genre clichés, making it a forgettable addition to the horror landscape."""
```

```python
prompt = f""" Classify the text into neutral, negative or positive.
Text: {TEXT}.
Sentiment:
"""
```

```python
sequences = pipe(
    prompt,
    max_new_tokens=30,
    # do_sample=True,
    return_full_text = False,
)
```

```python
for seq in sequences:
    print(f"Result: {seq['generated_text']}")
```
